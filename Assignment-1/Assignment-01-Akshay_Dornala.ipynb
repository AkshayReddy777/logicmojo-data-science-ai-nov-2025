#### Question-1

Write a function `second_largest(numbers)` that takes a list of numbers and returns the second largest number in the list. Assume the list has at least two unique numbers.

Example:
```python
numbers = [10, 5, 20, 8, 15]
second = second_largest(numbers)
print(second)
# Expected output: 15
```

````python
# Write your code here

def second_largest(numbers : list) -> int:
    '''
    This function returns the second largest number from a list of integers.
    Assumes that the list contains atleast two distinct integers, so no need for this error handling.
    List is converted into a set to remove duplicates and then second last element is returned.
    example:
    second_largest([1, 3, 5, 5, 5]) -> 3 
    '''
    result = sorted(set(numbers))[-2]
    #print(result)
    return result;

second_largest([1, -3, -15, 5, 5])

````

#### Question-3

Write a Python program to print all the prime numbers between 1 and N (inclusive). Where N is a positive integer and an input from user.

Example:

```python
N = 20
# Expected output: 2,3,5,7,11,13,17,19

```

````python
# Write your code here

def prime(limit : int):
    '''
    This function returns a list of all prime numbers upto a given limit including the limit.

    example:

    prime(7) -> {2, 3, 5, 7}    

    '''
    assert limit > 1, "Limit should be greater than 1"
    primes = []
    for n in range(2, limit + 1):
        is_prime = True
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break

        if is_prime:
            primes.append(n)

    return primes

print(prime(20))        


````

#### Question-6

Write a function `is_anagram(str1, str2)` that checks if two given strings are anagrams of each other (contain the same characters with the same frequencies, ignoring case).

Example:
```python
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
# Expected output:
# True
# False
```

````python
# Write your code here

def anagram(str1 : str, str2 : str) -> bool:
    '''
    This function takes two strings as input and returns True if they are anagrams of each other, False otherwise.

    example:
    "listen", "silent" -> True
    "hello", "world" -> False
    '''
    
    string1 = str1.lower().replace(" ", "").strip();
    string2 = str2.lower().replace(" ", "").strip();

    return sorted(string1) == sorted(string2);

print(anagram("Listen", "Silent"))
print(anagram("Hello", "World"))
print(anagram("Tri angle", "Int   egral"))

````

#### Question-7

Write a function `rotate_list(lst, k)` that rotates a given list lst to the right by `k` positions in-place. Assume `k` is a non-negative integer and can be greater than the length of the list.

Example:

```python
my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)
# Expected output: [4, 5, 1, 2, 3]
```

````python
# Write your code here

def rotate_list(lst : list, k : int) -> list:
    '''
    This function takes a list and an integer n as input and returns the list rotated to the right by n positions.

    example:
    [1, 2, 3, 4, 5], 2 -> [4, 5, 1, 2, 3]
    '''
    k = k % len(lst)
    rotated_list = lst[-k:] + lst[:-k]
    return rotated_list;

print(rotate_list([1, 2, 3, 4, 5], 7))
````

#### Question-8

You have a list of dictionaries, where each dictionary represents a product with keys "name", "price", and "category". Write a function group_by_category(products) that takes this list and returns a dictionary where the keys are the unique categories and the values are lists of product names belonging to that category.

Example:

```python
products = [
    {"name": "Laptop", "price": 1200, "category": "Electronics"},
    {"name": "Book", "price": 25, "category": "Books"},
    {"name": "Tablet", "price": 300, "category": "Electronics"},
    {"name": "Fiction", "price": 15, "category": "Books"},
]
grouped_products = group_by_category(products)
print(grouped_products)
# Expected output: {'Electronics': ['Laptop', 'Tablet'], 'Books': ['Book', 'Fiction']}

```


````python
# Write your code here
def group_by_category(products) -> dict:
    '''
    This function takes a list of tuples where each tuple contains a product name and its category.
    It returns a dictionary where the keys are categories and the values are lists of product names in that category.
    '''
    category_values = {}
    for i in products:
        name = i['name']
        category = i['category']
        if category in category_values:
            category_values[category].append(name)
        else:
            category_values[category] = [name]
    return category_values;

products = [
    {"name": "Laptop", "price": 1200, "category": "Electronics"},
    {"name": "Book", "price": 25, "category": "Books"},
    {"name": "Tablet", "price": 300, "category": "Electronics"},
    {"name": "Fiction", "price": 15, "category": "Books"},
]
print(group_by_category(products))
````

````python
# Write your code here

def pascal_triangle(n : int):
    '''
    This function prints the first n rows of Pascal's triangle.

    example:
    pascal_triangle(5) ->
        1
       1 1
      1 2 1
     1 3 3 1
    1 4 6 4 1
    '''
    for row in range(n):
        print(" " * (n - row), end="")
        value = 1
        for col in range(row + 1):
            print(value, end=" ")
            value = value * (row - col) // (col + 1)
        print()


pascal_triangle(5)        
````

#### Question-10

Write a function `reverse_the_words(s)` which will take a string `s` as input and reverses the words in the string. 

Example:

```python
s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)
# Expected output: challenging and awesome is Python
```

````python
# Write your code here
def reverse_the_words(s : str) -> str:
    '''
    This function takes a string as input and returns the string with the order of words reversed.

    example:
    "Hello World from Python" -> "Python from World Hello"
    '''
    words = s.split()
    reverse_word = ' '.join(reversed(words))
    return reverse_word;

reverse_the_words("Python is awesome and challenging")
````

#### Bonus question

Write a function `word_counter(s)` which will take a string `s` as input and returns a dictionary with count of words (case insensitive). Remove all the punctuations and special symbols present in the string.  

Example:
```python
s = "This city is a beautiful city. This is where I live"
d = word_counter(s)

# Expected output (output can be in any order): {"this": 2, "city": 2, "is": 2, "a": 1, "beautiful": 1, "where": 1, "i": 1, "live": 1}
```

````python
# Write your code here

def word_counter(s : str) -> dict:
    '''
    This function takes a string as input and returns a dictionary with the count of each word in the string.
    removes all punctuation and special symbols from the string.
    example:
    "This is a test. This test is only a test!" ->
    {
        "this": 2,
        "is": 2,
        "a": 2,
        "test": 3,
        "only": 1
    }
    '''
    import string
    s = s.translate(str.maketrans('', '', string.punctuation))
    s = s.lower()
    words = s.split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count;


word_counter("This is a test. This test is only a test!")
````
