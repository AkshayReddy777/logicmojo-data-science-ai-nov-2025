#### 11. Add a column that calculates the percent change between the 2010 Census and the 2017 Population Estimate

```python
# Write your code here

pop['Percentage change'] = ((pop['2017PopEstimate'] - pop['2010Census']) / pop['2010Census']) * 100

pop.head(10)
```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.

```python
# Write your code here

state_pop = (
    pop.groupby("State", as_index=False)[["2010Census", "2017PopEstimate"]]
      .sum()
)

state_pop['Percentage change'] = ((state_pop['2017PopEstimate'] - state_pop['2010Census']) / state_pop['2010Census']) * 100
state_pop.sort_values(by='Percentage change', ascending=False).head(5)
```
